```python
import pandas as pd

# Load CPI data, skipping the title and sub-header rows
cpi = pd.read_excel("CPI New.xlsx", sheet_name="Sheet2", header=2)

# Clean column names
cpi.columns = cpi.columns.str.strip().str.lower().str.replace(' ', '_')

# Print the column names after cleaning to verify
print(cpi.columns.tolist())

# Now, select and rename columns based on the verified column names
cpi = cpi[['all_itemsmonthly',  # CPI Index
    'all_items_month-on_change_(%)',  # Monthly inflation
    'all_items_year-on_change_(%)',   # Yearly inflation
    'food_monthly',
    'food_month-on_change_(%)',
    'food_year-on_change_(%)',
    'all_items_less_farm_produce_and_energy',
    'month-on_change_(%)',  # Core CPI monthly
    'year-on_change_(%)'    # Core CPI yearly
]].rename(columns={
    'all_itemsmonthly': 'cpi_index',
    'all_items_month-on_change_(%)': 'cpi_mom',
    'all_items_year-on_change_(%)': 'cpi_yoy',
    'food_monthly': 'food_index',
    'food_month-on_change_(%)': 'food_mom',
    'food_year-on_change_(%)': 'food_yoy',
    'all_items_less_farm_produce_and_energy': 'core_index',
    'month-on_change_(%)': 'core_mom',
    'year-on_change_(%)': 'core_yoy'
})
```
